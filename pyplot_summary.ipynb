Introduction for pyplot and matplotlib

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
def fig1():
    fig = plt.figure(num=1, figsize = [12,5])
    subplot = fig.add_subplot(121)
    subplot.set_title("Reflection Spectra", fontsize=18)
    subplot.set_xlabel('$\lambda $ (nm)', fontsize=14)
    subplot.set_ylabel("Reflectance (%)", fontsize=14)
    subplot.axis([0,1.2,-2,4])
    
    x_vec = np.linspace(0,1,100)
    subplot.plot(x_vec, np.sin(30*x_vec), color = 'cyan',linewidth=4,label="sine")
    
    subplot.legend(loc='upper left', bbox_to_anchor=(1,1), title="title", title_fontsize = 7 , fontsize = "large", shadow=True)

    subplot2 = fig.add_subplot(122)
    subplot2.set_title("Reflection Spectra", fontsize=18)
    subplot2.set_xlabel('$\lambda $ (nm)', fontsize=14)
    subplot2.set_ylabel("Reflectance (%)", fontsize=14)
    subplot2.axis([0,10,-2,4])
    
    data = [np.linspace(0,10,30),np.sin(np.linspace(0,10,30)),np.random.randn(30)] 
   
    subplot2.legend(loc='upper left', bbox_to_anchor=(1,1), title="title", title_fontsize = 12 , fontsize = "large", shadow=True)
    subplot2.text(0.5, 0.85, "Textbox", transform=subplot2.transAxes, fontsize=14)

    inset = fig.add_axes([0.4, 0.3, 0.2, 0.1])
    inset.plot(range(1000)[::-1], color='green')

    plt.tight_layout()
    
if __name__=='__main__':
    fig1()
```

```python
def fig2():
  
    data = [np.random.rand(200),np.random.rand(200),np.random.rand(200)]
   
    fig = plt.figure(2,[10,6])
    
    grp1 = fig.add_subplot(231)
    grp1.set_title("scatter")
    grp1.scatter(data[0],data[1],c=data[2], cmap = "RdBu_r")
  
    grp2 = fig.add_subplot(232)
    grp2.set_title("contour for non-uniform data")
    grp2.tricontourf(data[0],data[1],data[2], levels=14, cmap = "RdBu_r")
    
    grp3 = fig.add_subplot(233)
    grp3.set_title("contour for uniform data, with cbar")
    ct3 = grp3.contourf(np.linspace(0,1,10),np.linspace(0,1,10),np.random.normal(1,1,(10,10)), levels=14, cmap = "RdBu_r")
    cbar = fig.colorbar(ct3, ticks = [-1,0,3], orientation='horizontal')
  
    #3D scatter plot
    grp4 = fig.add_subplot(234, projection='3d')
    grp4.scatter(data[0],data[1],data[2])
    
    #3D Surface Plot (uniform data required)
    grp5 = fig.add_subplot(235, projection='3d')
    grp5.plot_surface(np.linspace(0,1,10),np.linspace(0,1,10),np.random.normal(1,1,(10,10)), cmap = "RdBu_r")
    
    #3D Vector plot
    vfdata = [np.random.rand(20),np.random.rand(20),np.random.rand(20),np.random.randn(20),np.random.randn(20),np.random.randn(20)]
    grp6 = fig.add_subplot(236, projection='3d')
    grp6.quiver(vfdata[0],vfdata[1],vfdata[2],vfdata[3],vfdata[4],vfdata[5], length = 0.1)
    plt.tight_layout()
    
if __name__=='__main__':
    fig2()
```
